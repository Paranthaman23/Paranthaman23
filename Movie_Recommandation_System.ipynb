**Import Dataset**

**Get Feature Selection**

```python
df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Director']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords'] + ' ' + df_features['Movie_Tagline'] + ' ' + df_features['Movie_Director']
```

```python
x
```

```python
x.shape
```

**Get features text conversation to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

**Get similarity Score using cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

**Get Movie name as a input**

```python
Favorited_Movie_Name = input('Enter your movie name: ')
```

```python
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommandation = difflib.get_close_matches(Favorited_Movie_Name, All_Movies_Title_List)
print(Movie_Recommandation)
```

```python
Close_Match = Movie_Recommandation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
```

```python
Recommadation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
print(Recommadation_Score)
```

```python
len(Recommadation_Score)
```

**Get All Movie Sort Based on Fav Movies**

```python
# sorting the movie with similarity score
sorted_similar_movies = sorted(Recommadation_Score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies
print('Top Movie Suggest for you : \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if (i<=30):
    print(i, '.',title_from_index)
    i+=1
```

**Top 10 Recommandation System**

```python
Movie_Name = input(' Enter your favourite movie name : ')
list_of_all_titles = df['Movie_Title'].tolist()
Find_close_match = difflib.get_close_matches(Movie_Name, list_of_all_titles)
Close_Match = Find_close_match[0]
Index_of_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recommadation_Score = list(enumerate(Similarity_Score[Index_of_Movie]))
sorted_similar_movies = sorted(Recommadation_Score, key = lambda x:x[1], reverse = True)
print('Top 10 Movies suggested for you : \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.Movie_ID==index]['Movie_Title'].values
  if (i<=10):
    print(i, '.',title_from_index)
    i+=1
```
